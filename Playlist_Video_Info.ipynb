```python
from tqdm import tqdm
```

```python
from pytube import Playlist

def get_playlist_videos_info(playlist_url):
    video_info_list = []  # Renamed list for clarity
    try:
        # Create a Playlist object
        playlist = Playlist(playlist_url)
        i=1
        # Fetch the title and duration of each video
        for video in tqdm(playlist.videos):
            try:
                title = video.title
                duration = video.length // 60  # Convert seconds to minutes
                # print(f"Title: {title}, Duration: {duration} minutes")
                video_info_list.append((i,title, duration))

            except Exception as video_error:
                # print(f"An error occurred while processing video: {video_error}")
                video_info_list.append((i,'error', 'na'))
            finally:
              i+=1

    except Exception as e:
        print(f"An error occurred: {e}")

    return video_info_list

video_info_list = get_playlist_videos_info(playlist_url)
```

```python
len(video_info_list)
```

```python
# .txt file
with open(subject+'_Lec_Info.txt', 'w') as file:
    for v in tqdm(video_info_list):
        s = f"{v[0]}: {v[1]}: {v[2]} minutes."
        file.write(s + '\n')
```

```python
# .csv file
with open(subject+'_Lec_Info.csv', 'w') as file:
    file.write("Index, Title, Duration" + '\n')
    for v in tqdm(video_info_list):
        v = [str(i).replace(',', '') for i in v]
        s = f"{v[0]}, {v[1]}, {v[2]}"
        file.write(s + '\n')
```
